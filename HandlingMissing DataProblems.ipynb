At first  glance, none of those variables has any missing values. However, looking into the data set carefully, missing values are indicated with zero. For Example: "SkinThickness" column with third row.


```python
data.head()
```

```python
for col in data.columns:
    print(data[col].isna().value_counts())
    
# only NaN valeus are ignored by operations like sum, count. 
```

For an efficient imputation and better counting, lets mark/convert these zeros to NaN value 


lets check how many zeros there are

```python
for cols in data.columns:
    print(cols, sum(data[cols]== 0))
```

```python
# replacing zero values with NaN but not for evey columns; for example, a pregnancy can be zero naturall 
# or we dont want to convert target value to NaN

data[data.columns[1:6]] = data[data.columns[1:6]].replace(0, numpy.NaN)
```

# 2 Imput Missing Values

Python's Pandas provides fillna() function for replacing missing values with statistical values such as mean. 

```python
#Pandas fillna() fucntion

data.fillna(value = data.mean())
```

```python
# scikit-learn also is a good too for imputation for missing values

from sklearn.preprocessing import Imputer
# we have to call and initiate Imputer function from sklearn library
imputer = Imputer() 
# as defeult Imputer uses mean as the imputation strategy 
trasformed_dataset = imputer.fit_transform(data[data.columns])


```
